```python
# 自动加入项目 src 到 sys.path
from dalong_catboost.notebook_utils import ensure_repo_in_sys_path
ROOT = ensure_repo_in_sys_path()
ROOT
```

```python
# 读取示例配置并展示关键字段
from dalong_catboost.config import load_config
import yaml

cfg = load_config(ROOT / "configs" / "example_binary.yaml")
cfg
```

```python
# 可选：尝试构建模型（无需真正训练）
from dalong_catboost.model import build_classifier
model = build_classifier(cfg.get("model", {}))
model
```

```python
import numpy
from catboost import CatBoostRegressor

dataset = numpy.array([[1,4,5,6],[4,5,6,7],[30,40,50,60],[20,15,85,60]])
train_labels = [1.2,3.4,9.5,24.5]
model = CatBoostRegressor(learning_rate=1, depth=6, loss_function='RMSE')
fit_model = model.fit(dataset, train_labels)

print(fit_model.get_params())
```
